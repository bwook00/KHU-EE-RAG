# Run AutoRAG

```python
from pathlib import Path
import os

root_dir = Path.cwd().resolve().parent.parent

qa_path = os.path.join(root_dir, 'data', 'qa', 'qa.parquet')
corpus_dir = os.path.join(root_dir, 'data', 'corpus', 'corpus_file_name.parquet')
project_dir = os.path.join(root_dir, 'results')
```

```python
import pandas as pd
qa = pd.read_parquet(qa_path)
qa
```

```python
corpus = pd.read_parquet(corpus_dir)
corpus
```

```python
from autorag.evaluator import Evaluator

evaluator = Evaluator(qa_data_path=qa_path,
                      corpus_data_path=corpus_dir,
                      project_dir=project_dir)
```

```python
yaml_dir = os.path.join(root_dir, 'config', 'config.yaml')
```

```python
evaluator.start_trial(yaml_dir)
```
